```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Load the shapefile (replace 'path_to_shapefile.shp' with the actual path)
shapefile_path = 'https://github.com/rcpsilva/UncertaintyQuantificationForGeocodingServices/raw/main/PaperGeoinfo/results/SMunicipios_2022/SP_Municipios_2022.shp'
gdf = gpd.read_file(shapefile_path)
```

```python
# Filter the data to select the specific city (e.g., São Paulo)
city_name = "São Paulo"  # Replace with the actual city name you want to plot
city_gdf = gdf[gdf['NM_MUN'] == city_name]
```

```python
import pandas as pd
```

```python
api = 'Google'
sp = pd.read_csv('geoGold_erro.csv')
sp = sp[sp['city']=='SAO PAULO']
sp = sp[sp['geoapi_id'] == api]
sp = sp[sp['erro']<0.1]
x = sp['lonGold']
y = sp['latGold']
erro = sp['erro']
```

```python
city_gdf.boundary.plot()
plt.plot(x,y,'k.')
plt.title('São Paulo City Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import numpy as np
```

```python
sp = pd.read_csv('geoGold_erro.csv')
sp = sp[sp['city']=='SAO PAULO']
global_min_error = np.min(sp['erro'])
global_max_error = 1 #np.max(sp['erro'])
max_erro = global_max_error
global_min_size = 1  # Minimum marker size
global_max_size = 20  # Maximum marker size
```

```python
api = 'MapBox'
sp = pd.read_csv('geoGold_erro.csv')
sp = sp[sp['city']=='SAO PAULO']
sp = sp[sp['geoapi_id'] == api]
sp = sp[sp['erro']<max_erro]
x = sp['lonGold']
y = sp['latGold']
erro = sp['erro']
marker_size = (erro - global_min_error) / (global_max_error - global_min_error) * (global_max_size - global_min_size) + global_min_size

city_gdf.boundary.plot()
plt.scatter(x, y, c=erro, cmap='viridis',alpha=1,s=marker_size,vmin=global_min_error, vmax=global_max_error) 
plt.colorbar()
plt.title(api)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
