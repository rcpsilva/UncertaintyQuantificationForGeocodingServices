```python
import pickle
import pandas as pd

n = 1000
```

## Get test data

```python
df = pd.read_csv('modeloJoinville.csv')

df = df.dropna()
df = df.sample(n = n)
```

```python
df.head()

X_test = df[['lat','lon']]
y_lon_test = df['lon_api']
y_lat_test = df['lat_api']
```

## Load models

```python
# Load data from the pickle file
with open('models_10000.pkl', 'rb') as file:
    regressorLat,regressorLon = pickle.load(file)
```

```python
lat_pred = regressorLat.predict(X_test.to_numpy())
```

```python
lon_pred = regressorLon.predict(X_test.to_numpy())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(lat_pred,y_lat_test,'o')
y_l,y_u = plt.ylim()
plt.xlim(y_l,y_u)
print(y_l,y_u)
```

```python
plt.plot(lon_pred,y_lon_test,'o')
y_l,y_u = plt.ylim()
plt.xlim(y_l,y_u)
print(y_l,y_u)
```
